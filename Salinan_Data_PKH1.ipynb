### **KLASIFIKASI KELAYAKAN PENERIMAAN PKH**

Program Keluarga Harapan (PKH) adalah program bantuan sosial yang diberikan kepada keluarga miskin untuk meningkatkan kesejahteraan mereka. PKH merupakan program pemberian uang tunai bersyarat yang disalurkan kepada Keluarga Penerima Manfaat (KPM). Tujuan PKH adalah: Mengurangi angka kemiskinan, Memutus mata rantai kemiskinan antar-generasi, Meningkatkan kualitas sumber daya manusia, Mengubah perilaku yang kurang mendukung dalam peningkatan kesejahteraan. PKH memberikan dukungan di berbagai sektor, seperti pendidikan, kesehatan, gizi, serta kebutuhan dasar lainnya. Peserta PKH diwajibkan melaksanakan persyaratan dan ketentuan yang telah ditetapkan, seperti kehadiran di fasilitas pendidikan atau fasilitas kesehatan.

```python
# 1. Mengatasi Missing Values
# Cek apakah ada nilai yang hilang
print("Cek Missing Values:")
print(df.isnull().sum())

# Jika ada nilai yang hilang, bisa diisi dengan median atau modus, atau hapus baris yang kosong
df.dropna(inplace=True)  # Menghapus baris dengan nilai NaN

# 2. Membersihkan Kolom Teks
df['Nama'] = df['Nama'].str.strip()
df['Keterangan'] = df['Keterangan'].str.strip()

# 3. Memastikan Tipe Data yang Benar
df['Pendapatan'] = pd.to_numeric(df['Pendapatan'], errors='coerce')
df['Umur'] = pd.to_numeric(df['Umur'], errors='coerce')

# 4. Mengatasi Duplikasi Data
# Cek duplikasi
print("\nCek Duplikasi Data:")
print(df.duplicated().sum())

# Menghapus baris duplikat jika ada
df.drop_duplicates(inplace=True)

# 5. Menangani Outliers pada Kolom Pendapatan menggunakan metode IQR
q1 = df['Pendapatan'].quantile(0.25)
q3 = df['Pendapatan'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Filter outliers
df = df[(df['Pendapatan'] >= lower_bound) & (df['Pendapatan'] <= upper_bound)]

# 6. Menambahkan Kolom 'Lansia' untuk menandai apakah usia >= 60 tahun
df['Lansia'] = df['Umur'] >= 60

# 7. Normalisasi Kolom Pendapatan (jika diperlukan)
df['Pendapatan_Normalized'] = (df['Pendapatan'] - df['Pendapatan'].min()) / (df['Pendapatan'].max() - df['Pendapatan'].min())

# 8. Menyimpan Data yang Bersih ke File CSV
df.to_csv("data_bersih.csv", index=False)

# 9. Visualisasi Distribusi Pendapatan
plt.figure(figsize=(10, 6))
sns.histplot(df['Pendapatan'], bins=10, kde=True)
plt.title("Distribusi Pendapatan")
plt.xlabel("Pendapatan")
plt.ylabel("Frekuensi")
plt.grid(True)
plt.show()

# Menampilkan Data yang Sudah Dibersihkan
print("\nData yang Sudah Dibersihkan:")
print(df)
```
